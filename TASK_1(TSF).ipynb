# Prediction Using Superwised Learning

Predict the % of students on the no. of study hours

```python
#importing all libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Reading the dataset("http://bit.ly/w-data")
df=pd.read_csv("https://bit.ly/w-data")
```

```python
#printing first rows of dataset 
df.head()
```

```python
#printing the shape of df
df.shape
```

```python
df.info()
```

```python
#plotting dataset given
df.plot(x='Hours',y='Scores',style='o')
plt.title('Hours vs Percentage')
plt.xlabel('percentage score')
plt.show()
```

```python
#seperating dependent and independent variable
x=df.drop(['Scores'],axis=1)
y=df['Scores']
```

```python
#creating training and testing data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
#fitting the model
re=LinearRegression()
re.fit(train_x,train_y)
```

```python
#printing the coefficient and intercept
print(re.coef_)
print(re.intercept_)
```

```python
#plotting of Regression line
line=re.coef_*x+re.intercept_

plt.scatter(x,y)
plt.plot(x,line,color='green');
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#predicting y value for 9.25 hours
hrs=9.25
pred_score=re.predict([[hrs]])
print('percentage scored is ',pred_score)
```
